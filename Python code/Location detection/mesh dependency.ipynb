# Robustness

```python
# Parameter 
nb_samples = 100000
batch = 1000

np.set_printoptions(suppress=True)

Meshs_train = [ "mesh_B" ]
Architectures = [build_model_24, build_model_40, build_model_60, build_model_84 ]
Data_sets = [ "data_set_p_2_all.csv","data_set_p_3_all.csv","data_set_p_4_all.csv","data_set_p_5_all.csv" ]
table_name = [ "Robustness_2.csv" ]
sigmas = [ 0 ,0.01, 0.05 , 0.1 ]
```

```python
Latex_table = []
for index , data_set in enumerate(Data_sets) :
    print("=============================================================================================")
    print(" the dataset is " , data_set)
    for mesh in Meshs_train:
        
        Latex_mesh = []
        
        models = []
        X_tests = []
        Y_tests = []
        for rounds in range(3):
            
            print(" ======== Train on " ,mesh,"============" )
            data_path = 'C:\\Users\\pierr\\Desktop\\Projet de semestre I\\data\\task_2\\train_set\\discontinuity_R\\'+mesh+'\\'+data_set

            # load the data 
            data = np.loadtxt(data_path, delimiter=',')
            data_sample = data_cut( data , nb_samples)
            # to make every model the same number of train and test in order to compare
            X_train, X_test_, Y_train, Y_test_, _, _ = data_preparation_approach_2(data_sample)
            model_ = Architectures[index]()
            train_mse,test_mse, results_angle ,results_dist = model_training(model_,X_train,X_test_,Y_train,Y_test_, patience_ = 100 ,batch_size_=batch, plot=True ,approach=2)
            models.append(model_)
            X_tests.append(X_test_)
            Y_tests.append(Y_test_)
        
        for sig in sigmas:
            rounds_results = []
            for ind_m , model in enumerate(models):
                if sig == mesh:
                    
                    Y_pred = model.predict(X_tests[ind_m] , verbose = 0 )
                    true_angle = np.arctan2(Y_tests[ind_m][:,0] ,Y_tests[ind_m][:,1])
                    pred_angle = np.arctan2(Y_pred[:,0 ] ,Y_pred[:,1] )
                    results_angle = diagnostic_angle(pred_angle , true_angle )
                    mse_2_angle = compute_loss_modify(pred_angle , true_angle)
                    print('MSE 2 angle test: %.3f' % (mse_2_angle))
                    
                    dist_pred = Y_pred[:,-1]
                    dist_true = Y_tests[ind_m][:,-1]
                    results_dist = diagnostic_distance(dist_pred, dist_true )
                    
                else :
                    X_noise = X_tests[ind_m] + np.random.normal(scale=sig , size=((X_tests[ind_m]).shape[0], (X_tests[ind_m]).shape[1]))
                    Y_pred = model.predict(X_noise , verbose = 0 )
                    true_angle = np.arctan2(Y_tests[ind_m][:,0] ,Y_tests[ind_m][:,1])
                    pred_angle = np.arctan2(Y_pred[:,0 ] ,Y_pred[:,1] )
                    results_angle = diagnostic_angle(pred_angle , true_angle )
                    mse_2_angle = compute_loss_modify(pred_angle , true_angle)
                    print('MSE 2 angle test: %.3f' % (mse_2_angle))
                    
                    dist_pred = Y_pred[:,-1]
                    dist_true = Y_tests[ind_m][:,-1]
                    results_dist = diagnostic_distance(dist_pred, dist_true )
                    
                rounds_results.append([ round(mse_2_angle,4), round(results_angle[0],4), round(results_angle[1],4), round(results_angle[2],4) , round(results_angle[4],4), round(results_angle[7],4),round(results_dist[0],4) ])
            
            Latex_mesh.append(np.mean(rounds_results,axis=0))
        Latex_table.append(Latex_mesh)   
        print(" Finish ALL test ")
        print("=======================================================================================================")
table = ((np.asarray(Latex_table)).swapaxes(1,2)).reshape((24,4))
np.savetxt(table_name[index], table, delimiter=',',fmt='%.3f')
```
