# Robustness 

```python
# Parameter 
nb_samples = 100000
modifie_angle = True 
modifie_loss = True
batch = 1000
All_results = []
All_train_mse = []
All_test_mse = []
save_model = []
All_who = []

Meshs_train = [ "mesh_B" ]
Meshs = [ "mesh_B" ]
Architectures = [build_model_24_cust_loss_sig ,build_model_40_cust_loss_sig,build_model_60_cust_loss_sig,build_model_84_cust_loss_sig ]
Data_sets = [ "data_set_p_2_all.csv","data_set_p_3_all.csv","data_set_p_4_all.csv","data_set_p_5_all.csv" ]
table_name = [ " robustness"]
sigmas = [ 0 ,0.01, 0.05 , 0.1 ]
np.set_printoptions(suppress=True)
```

```python
Latex_table = []
for index , data_set in enumerate(Data_sets) :
    print("=============================================================================================")
    print(" the dataset is " , data_set)
    
    for mesh in Meshs_train:
        
        Latex_mesh = []
        
        models = []
        X_tests = []
        Y_tests = []
        for rounds in range(3):
            
            print(" ======== Train on " ,mesh,"============" )
            data_path = 'C:\\Users\\pierr\\Desktop\\Projet de semestre I\\data\\train_set\\discontinuity_R\\'+mesh+'\\'+data_set

            # load the data 
            data = np.loadtxt(data_path, delimiter=',')
            data= data[:,1:]
            data_sample = data_cut( data , nb_samples)
            # to make every model the same number of train and test in order to compare
            X_train, X_test_, Y_train, Y_test_, indice_neg_data = data_preparation(data_sample,modifie_angle)
            model_, train_mse , test_mse,results = training_NN(Architectures[index],X_train, X_test_, Y_train, Y_test_,patience_ = 100,batch = batch,modifie_loss =modifie_loss )
            models.append(model_)
            X_tests.append(X_test_)
            Y_tests.append(Y_test_)
        
        for sig in sigmas:
            rounds_results = []
            for ind_m , model in enumerate(models):
                if sig == 0:
                    y_pred = model.predict(X_tests[ind_m],verbose=0)
                    results = diagnostic(y_pred , Y_tests[ind_m],  True , modifie_loss )
                    test_mse = model.evaluate(X_tests[ind_m],Y_tests[ind_m], verbose=0)
                    print("on a new mesh the loss is :",format(test_mse,'.4f'))
                else:
                    X_noise = X_tests[ind_m] + np.random.normal(scale=sig , size=((X_tests[ind_m]).shape[0], (X_tests[ind_m]).shape[1]))
                    y_pred = model.predict(X_noise,verbose=0)
                    results = diagnostic(y_pred , Y_tests[ind_m],  True , modifie_loss )
                    test_mse = model.evaluate(X_noise,Y_tests[ind_m], verbose=0)
                    print("on a new mesh the loss is :",format(test_mse,'.4f'))
                rounds_results.append([ round(test_mse,4), round(results[0],4), round(results[1],4), round(results[4],4) , round(results[2],4), round(results[7],4) ])

            Latex_mesh.append(np.mean(rounds_results,axis=0))
        Latex_table.append(Latex_mesh)   
        print(" Finish ALL test ")
        print("=======================================================================================================")
table = ((np.asarray(Latex_table)).swapaxes(1,2)).reshape((24,4))
np.savetxt(table_name[0], table, delimiter=',',fmt='%.3f')
```
