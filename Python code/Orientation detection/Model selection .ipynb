# Model selection 
we will test a few possible architecture in order to select to the best one 

```python
from fonction import *
from models import *
```

```python
# Parameter 
nb_samples = 100000
modifie_angle = True 
modifie_loss = True
batch = 1000
All_results = []
All_train_mse = []
All_test_mse = []
save_model = []
All_who = []

np.random.seed(1234)

Meshs = [ "mesh_A" ,"mesh_B" ,"mesh_C"]
Architectures = [build_model_40 , build_model_40_cust_loss , build_model_40_cust_loss_relu ,
                 build_model_40_cust_loss_wider , build_model_40_cust_loss_deeper , build_model_40_cust_loss_deeper_2 ,
                build_model_40_cust_loss_batch , build_model_40_cust_loss_deeper_batch, build_model_40_cust_loss_wider_batch,
                build_model_40_cust_loss_deeper_2_batch,build_model_40_cust_loss_batch_sig,
                build_model_40_cust_loss_sig]
Data_sets = [ "\\data_set_p_3_all.csv" ]
```

###### Custom loss 

```python
for mesh in Meshs :
    print("=============================Model==================================")
    print(mesh)
    print(Data_sets[0])
    All_who.append(mesh+Data_sets[0])
    data_path = 'C:\\Users\\pierr\\Desktop\\Projet de semestre I\\data\\train_set\\discontinuity_R\\'+mesh+Data_sets[0]

    # load the data 
    data = np.loadtxt(data_path, delimiter=',')
    ind = data[:,0]
    data= data[:,1:]
    data_sample = data_cut( data , nb_samples)
    # to make every model the same number of train and test in order to compar
    X_train, X_test, Y_train, Y_test, indice_neg_data = data_preparation(data_sample,modifie_angle)
    model, train_mse , test_mse,results = training_NN(Architectures[1],X_train, X_test, Y_train, Y_test,patience_ = 100,batch = batch,modifie_loss =modifie_loss )
    save_model.append(model)
    All_results.append(results)
    All_train_mse.append(train_mse)
    All_test_mse.append(test_mse)
```
