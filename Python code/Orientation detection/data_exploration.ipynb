### Remove negative angle
Indead and orientation can be define by two angles , so we can keep the one that is positive 

```python
def angle_visualization_output(Y_test, nbr_of_plot = 10):
    
    pts = np.array([[-1,-1],[-1,1],[1,-1]])
    for i in range(nbr_of_plot):
        plt.figure()
        triangle = plt.Polygon(pts,fill=False)
        plt.gca().add_patch(triangle)
        plt.arrow(-0.5,0,np.cos(Y_test[i]),np.sin(Y_test[i]),head_width=0.05, head_length=0.1,color='r', label =' true ')
        plt.xlim(-1.5,1.7)
        plt.ylim(-1.5,1.5)
        plt.title(" Angle visualization the ange is : " + str(Y_test[i])+" radian and "+ str(Y_test[i]*180/np.pi))
        plt.show()
```

```python
random_permuted_indices = np.random.permutation(np.shape(indice_neg)[0])
angle_visualization_output(Y[indice_neg[random_permuted_indices]])
```

```python
angle_visualization_output(Y[indice_neg[random_permuted_indices]]+np.pi)
```
